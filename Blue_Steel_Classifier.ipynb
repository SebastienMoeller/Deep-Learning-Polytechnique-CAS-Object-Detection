## Start from here

```python
class classifier(nn.Module):
    def __init__(self, num_of_class, dropout_rate):
        super(classifier, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=2),
            nn.Dropout2d(dropout_rate),
            nn.ReLU(),
            nn.BatchNorm2d(16),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=2),
            nn.Dropout2d(dropout_rate),
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d(kernel_size=2, stride=2))
        #self.fc = nn.Linear(32 * 56 * 56, num_of_class)
        self.fc = nn.Linear(32 * 7 * 7, num_of_class)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        #print(out.shape)
        #out = out.reshape(-1, 32 * 56 * 56)
        out = out.reshape(-1, 32 * 7 * 7)
        #out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        #return F.log_softmax(out, dim = 1)
        return out
```

```python
print(classifier(NUM_CLASS, DROPOUT_RATE))
```

```python
def train(model, data_loader, loss_fn, optimizer, n_epochs=1):
    
    model.train(True)
    loss_train = np.zeros(n_epochs)
    acc_train = np.zeros(n_epochs)
    loss_test = np.zeros(n_epochs)
    acc_test = np.zeros(n_epochs)
    
    for epoch_num in range(n_epochs):
        running_corrects = 0.0
        running_loss = 0.0
        size = 0
        
        for i, data in enumerate(data_loader, 0):
            image, label = data
            label, image = label.to(device), image.to(device)
            output = model.forward(image)
            loss = loss_fn(output, label)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            pred = output.max(1, keepdim=True)[1]
            running_corrects += pred.eq(label.view_as(pred)).sum()
            running_loss += loss
            size += label.size(0)
            
        epoch_loss = running_loss / size
        epoch_acc = running_corrects.item() / size
        loss_train[epoch_num] = epoch_loss
        acc_train[epoch_num] = epoch_acc
        loss_test[epoch_num], acc_test[epoch_num] = test(model, test_loader, loss_fn)
        
        if epoch_num < 9:
            print('Epoch [{}/{}]  | Train - Loss: {:.4f} Acc: {:.2f}% | Test - Loss: {:.4f} Acc: {:.2f}%'.format(epoch_num+1, n_epochs, epoch_loss, epoch_acc*100, loss_test[epoch_num], acc_test[epoch_num]*100))
        else:
            print('Epoch [{}/{}] | Train - Loss: {:.4f} Acc: {:.2f}% | Test - Loss: {:.4f} Acc: {:.2f}%'.format(epoch_num+1, n_epochs, epoch_loss, epoch_acc*100, loss_test[epoch_num], acc_test[epoch_num]*100))

    return loss_train, acc_train, loss_test, acc_test
```

```python
def test(model, data_loader, loss_fn):
    with torch.no_grad():
        running_correct = 0.0
        running_loss = 0.0
        size = 0
        
        for i, data in enumerate(data_loader, 0):
            image, label = data
            image = image.to(device)
            label = label.to(device)
            outputs = model(image)
            
            loss = loss_fn(outputs, label)
            predicted = torch.argmax(outputs,dim=1)
            
            size += label.size(0)
            running_correct += (predicted == label).sum().item()
            running_loss += loss
        loss_test = running_loss / size
        acc_test = running_correct / size
    return loss_test, acc_test
```

```python
%%time

DROPOUT_RATE = 0

if __name__ == '__main__':
    model0 = classifier(NUM_CLASS, DROPOUT_RATE).to(device)
    optimizer0 = torch.optim.Adam(model0.parameters(), lr = LR)
    loss_fn0 = nn.CrossEntropyLoss()
    loss_train0, acc_train0, loss_test0, acc_test0 = train(model0, train_loader, loss_fn0, optimizer0, TRAIN_EPOCH)
```
